<h2> Manipulando os Dados </h2>
<h3> Propósito: gerar uma planilha em que cada linha se refere a um paciente. </h3>
<h3>Possíveis Problemas: dados repetidos, vacinas diferentes pra mesma pessoa,outras    
</h3>

```python
#Dicionário guardará a lista de informações de cada paciente
dic = {}

#remove duplicatas
data_df.drop_duplicates(inplace=True)

#salva os dados
for index, row in data_df.iterrows():
    temp = dic.get(row['paciente_id'],False)
    if temp==False:
        dic[row['paciente_id']]=list()
        dic[row['paciente_id']].append(row['paciente_id'])
        dic[row['paciente_id']].append(row['vacina_nome'])
        dic[row['paciente_id']].append(row['paciente_racacor_valor'])
        dic[row['paciente_id']].append(row['paciente_idade'])
        dic[row['paciente_id']].append(row['paciente_enumsexobiologico'])
        dic[row['paciente_id']].append(row['estabelecimento_municipio_nome'])
        dic[row['paciente_id']].append(row['estabelecimento_uf'])
        #remove datas repetidas 
    if not(dic[row['paciente_id']][-1]==row['vacina_dataaplicacao'] or dic[row['paciente_id']][-2]==row['vacina_dataaplicacao'] ):
        dic[row['paciente_id']].append(row['vacina_dataaplicacao'])
```

```python
#Passamos os dados para uma lista pra facilitar a transformação em planilha
full = []
for ids in dic.keys():
    if len(dic[ids])>9:
        print(dic[ids])
    full.append(dic[ids])
```

```python
from pandas import DataFrame
dt=DataFrame(full,columns=['paciente_id','vacina_nome','paciente_racacor_valor','paciente_idade','paciente_enumsexobiologico','estabelecimento_municipio_nome','estabelecimento_uf','vacina_dataaplicacao_primeira_dose','vacina_dataaplicacao_segunda_dose','d3','d4','d5'])
dt.to_csv("output.csv")  
```
